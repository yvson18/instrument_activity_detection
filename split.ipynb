```python
import os
import subprocess
import random
```

```python
file_path= [os.path.join("full_dataset", arquivo) for arquivo in os.listdir("full_dataset") if os.path.isfile(os.path.join("full_dataset", arquivo))]
```

```python
random.seed(21)
random.shuffle(file_path)

#create train, validation and test sets

train_set = file_path[:int(len(file_path)*0.8)]
validation_set = file_path[int(len(file_path)*0.8):int(len(file_path)*0.9)]
test_set = file_path[int(len(file_path)*0.9):]
```

```python
type_split = {"train": train_set, "validation": validation_set, "test": test_set}
             
for type_ in list(type_split.keys()):
    save_path = f"full_dataset_wav/{type_}"

    for file in type_split[type_]:
        saving_path = [
            save_path + "/mixture/" + file.replace("full_dataset/","").replace(".mp4", ".wav"),
            save_path + "/drums/" + file.replace("full_dataset/","").replace(".mp4", ".wav"),
            save_path + "/bass/" + file.replace("full_dataset/","").replace(".mp4", ".wav"),
            save_path + "/rest/" + file.replace("full_dataset/","").replace(".mp4", ".wav"),
            save_path + "/vocals/" + file.replace("full_dataset/","").replace(".mp4", ".wav")
        ]
        #print(saving_path)
        for i in range(len(saving_path)):
            comando = ['ffmpeg', '-i', file, '-map', '0:{0}'.format(i), saving_path[i]]
            subprocess.call(comando)
```
